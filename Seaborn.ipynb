# Introduction to Seaborn

## Seaborn library : 

### - Used for data visualization & based on Matplotlib
### - Seaborn allows the creation of statistical grpahics.

## Functionalities:

### - Allows comparison between multiple variables
### - Suppports multi-plot grids
### - Available univariate and bivariate visualizations
### - Available of different colour palettes
### - Estimates and plots linear regression automatically 

Seaborn fixes two shortcomings of Matplotlib:
   
     Matplotlib settings are difficult to figure out, Seaborn comes with numerous 
     customized themes ang high-level interfaces.
    
    Matplotlib doesn't serve well when it comes to dealing with DataFrames, while 
    Seaborn functions actually work on DataFrames.

# Visualizing Statistical Relationships
## process of understanding relationships between variables of a dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = sns.load_dataset("flights")
```

```python
sns.relplot(x = "passengers",y = "month", data = a)
```

```python
sns.relplot(x = "passengers",y = "month",hue = "year", data = a)
```

```python
b = sns.load_dataset("tips")
```

```python
sns.relplot(x= "time", y="tip", data = b, kind = "line")
```

# Plotting with Categorical Data

Main variable is further divided into discrete groups

```python
sns.catplot(x="day", y="total_bill",data = b)
```

```python
sns.catplot(x="day", y="total_bill",kind= "violin",data = b)
```

```python
sns.catplot(x="day", y="total_bill",kind= "boxen",data = b)
```

# Multi- PLot grids

Graphs are plotted side-by-side using the same scale and axes to aid comparison 

```python
a= sns.load_dataset("iris")
```

```python
b = sns.FacetGrid(a,col="species")
b.map(plt.hist, "sepal_length")
```

```python
a=sns.load_dataset("flights")
```

```python
b=sns.PairGrid(a)
b.map(plt.scatter)
```

# Plot- Aesthetics

## Deal with making our plots more attractive and delightful: using various themes and colo-palettes


```python
sns.set(style= "dark")
b=sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style= "white",color_codes=True)
a= sns.load_dataset("tips") 
sns.boxplot(x="day", y="total_bill",data=a)
```

```python
sns.set(style= "white",color_codes=True)
a= sns.load_dataset("tips") 
sns.boxplot(x="day", y="total_bill",data=a)
sns.despine(offset=10, trim = True)
```

```python
c = sns.color_palette()
sns.palplot(c)
```
